```python
discounts = pd.read_excel(r'D:\Yuri\Projects\DSS_169_increased_marketplaces_discount\final_discounts.xlsx')
discounts
```

```python
countries_df = discounts.groupby(['ts_spawn_date', 'country'], as_index=False).agg({'lead_qty': sum})
countries_df = countries_df.query('country == ["Chile", "Mexico", "Peru"]')
countries_df = countries_df.head(84)
countries_df = countries_df.query('ts_spawn_date != ["2022-08-02", "2022-08-08", "2022-08-09", "2022-08-10", "2022-08-18", "2022-08-21", "2022-08-23"]')
```

```python
result_list = []
for i in countries_df['country'].unique():
    test = countries_df[(countries_df['country'] == i)]
    group1 = test[['ts_spawn_date', 'country', 'lead_qty']]
    group1 = group1.query('ts_spawn_date < "2022-08-16"')
    group2 = test[['ts_spawn_date', 'country', 'lead_qty']]
    group2 = group2.query('ts_spawn_date > "2022-08-15"')
    result_list.append(ttest_rel(group1['lead_qty'], group2['lead_qty']))
```
